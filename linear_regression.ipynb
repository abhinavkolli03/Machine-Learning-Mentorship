```
learn_rate = .00001
final_da = 0
final_db = 0
loss = 0
def update(final_da, final_db):
    da = 0
    db = 0
    for i in range(0, M):           
        da += 2 * (float(final_da) * float(x_values[i]) + float(final_db) - float(y_values[i])) * float(x_values[i]) 
        da = da/M 
        db += 2 * (float(final_da) * float(x_values[i]) + float(final_db) - float(y_values[i]))
        db = db/M 
        #Both sum formulas of regression errors
    final_da = final_da - learn_rate*da #update da
    final_db = final_db - learn_rate*db #update db

txt = open('data.train', 'r')
M = 0
x_values = []
y_values = []
for line in txt:
    pt = line.split(',')
    x_values.append(pt[0])
    y_values.append(pt[1])
    M += 1

for i in range (0, M):
    previous = loss
    loss = 0
    for j in range (0, M):
        loss += (float(final_da)*float(x_values[j]) + float(final_db) - float(y_values[j]))**2
    loss = loss/M
    if loss > previous:
        print(f'f(x) = {final_da}x + {final_db}.\n Loss = {loss}')
        break
    else:
        update(final_da, final_db)
```
