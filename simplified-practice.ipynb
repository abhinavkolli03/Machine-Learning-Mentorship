```
#recursion for a factorial
def fact(n):
  #base case
  if n == 1:
    return n

  #logic
  return n * fact(n-1)
n = int(input())
if n < 0:
  print("Not Possible")
else:
  print(fact(n))
```

```
#recursion for a fibonacci
def fibonacci(n):
  if n <= 1:
    return n
  return fibonacci(n-1) + fibonacci(n-2)

n = int(input())
if n <= 0:
  print("not possible")
else:
  for i in range (n):
    print(fibonacci(i))
```
