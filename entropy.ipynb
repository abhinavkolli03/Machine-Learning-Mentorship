```
#Inheritance
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

class Student(Person):
  def __init__(self, fname, lname, year):
    super().__init__(fname, lname) #calls the Parent _init_
    self.graduationyear = year

x = Student("Mike", "JJ", 2019)
y = Person("Ashutosh", "Senapati")
x.printname()
y.printname()
```
