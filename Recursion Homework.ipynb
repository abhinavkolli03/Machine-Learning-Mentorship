Question 2

```
def fib(n):
  if n <= 1:
    return n
  return fib(n-1) + fib(n-2)
n = int(input())
if n <= 0:
  print("Not Possible")
else:
    print(fib(n))
#Since higher numbers require more recursions and call methods, time complexity will be bigger.
```

```
def main():
  print(digitvalue(15))
  print(digitvalue(105))
  print(digitvalue(15105))
def digitvalue(num):
  if num < 10:
    return 1
  return 1 + digitvalue(num//10)
main()
```

```
x = [1, 10, 15, 281, 2304, 3490]
def findmax(a):
  b = a[1:len(a)]
  if (len(a)) == 1:
    return a[0]
  else:
    return max(a[0],findmax(b))
print(findmax(x))
```
