```
#TicTacToe Example
gameboard = [["-", "-", "-"], 
             ["-", "-", "-"], 
             ["-", "-", "-"]]

#updates elements in gameboard
def update_gameboard(ps):
  if(ps[0] == "x"):
    gameboard[int(ps[1])][int(ps[2])] = "X"
  else:
    gameboard[int(ps[1])][int(ps[2])] = "O"
  return gameboard

#checks if winner or draw
def check_winner():
  complete = False
  #rows check
  for i in range (len(gameboard)):
    if(gameboard[i][0] == "X" and gameboard[i][1] == "X" and gameboard[i][2] == "X"):
      complete = True
      print("Game over: Player X wins!")
    elif(gameboard[i][0] == "X" and gameboard[i][1] == "X" and gameboard[i][2] == "X"):
      complete = True
      print("Game over: Player Y wins!")
  #columns check
  for i in range (len(gameboard[0])):
    if(gameboard[0][i] == "X" and gameboard[1][i] == "X" and gameboard[2][i] == "X"):
      complete = True
      print("Game over: Player X wins!")
    elif(gameboard[0][i] == "O" and gameboard[1][i] == "O" and gameboard[2][i] == "O"):
      complete = True
      print("Game over: Player Y wins!")
  #diagonals check
  if(gameboard[0][0] == "X" and gameboard[1][1] == "X" and gameboard[2][2] == "X"):
      complete = True
      print("Game over: Player X wins!")
  elif(gameboard[0][0] == "O" and gameboard[1][1] == "O" and gameboard[2][2] == "O"):
      complete = True
      print("Game over: Player Y wins!")
  if(gameboard[2][0] == "X" and gameboard[1][1] == "X" and gameboard[0][2] == "X"):
      complete = True
      print("Game over: Player X wins!")
  elif(gameboard[2][0] == "O" and gameboard[1][1] == "O" and gameboard[0][2] == "O"):
      complete = True
      print("Game over: Player Y wins!")
  #final check
  dash_checker = True
  for i in range (len(gameboard)):
    for j in range (len(gameboard[0])):
      if gameboard[i][j] == "-":
        dash_checker = False
  if dash_checker and complete == False:
    complete = True
    print("Game over: Draw")
  return complete

#prints the board
def print_board(gameboard):
  for i in range(len(gameboard)):
      print(str(gameboard[i]) + " ")
  print("\n")

#main method
turn_count = 1
while(check_winner() == False):
    if(turn_count % 2 == 1):
      player_turn = input("Player X: please input your turn - player pos1 pos2 (EX: x 0 0) >>> ")
    else:
      player_turn = input("Player Y: please input your turn - player pos1 pos2 (EX: y 0 0) >>> ")
    ps = player_turn.split()
    gameboard = update_gameboard(ps)
    print_board(gameboard)
    turn_count += 1
```
